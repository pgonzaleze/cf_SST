## Code to get all the links in a page
## Create a list with all the URL
### Read each of the URL´s to download the NetCDF file

```python
## To get all the links in a page
## Create a list with all the URL
### Read each of the URL´s to download the NetCDF file

from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

import requests
r  = requests.get("https://www.ncei.noaa.gov/data/avhrr-reflectance-cloud-properties-patmos-extended/access/2018/")
data = r.text
soup = BeautifulSoup(data)

URL_list=[]
for link in soup.find_all('a'):
    URL = (link.get('href'))
    URL_list.append(URL)
    
```

```python
len(URL_list)
```

```python
#Clean URL list to get only 'asc' files
URL_asc=[]
for TextLine in URL_list:
    if '_asc_' in TextLine:
        FoundLine = TextLine
        URL_asc.append(FoundLine)
```

```python
len(URL_asc)
```

## Select specific satellite in case that you have to select two different satellites< for exaple 1985 uses NOAA 7 and 9

```python
## select specific satellite in case that you have to select two different satellites< for exaple 1985 uses NOAA 7 and 9
URL_NOAA18=[]
for TextLine in URL_asc:
    if '_NOAA-18_' in TextLine:
        FoundLine = TextLine
        URL_NOAA18.append(FoundLine)
```

## Loop through the 'URL_asc' list to download the files

```python
# Loop through the 'URL_asc' list to download the files
# do not forget to change the correct "satellite list" (URL_asc, URL_NOAA...) 
for yearcount,year in enumerate(np.arange(2018,2019)):
    if yearcount > 40:
        break
    for line in (URL_NOAA18):
        root = "https://www.ncei.noaa.gov/data/avhrr-reflectance-cloud-properties-patmos-extended/access/" +'{}'.format(year)
        filename = root + '/'+'{}'.format(line)
        download(filename)
        #print(filename)
        time.sleep(3)
    print('-'*20)
```
